```python
#5 annotations
#---visit this link for the description of the dataset:
#---https://archive.ics.uci.edu/ml/datasets/Air+Quality

#---data preparation


df_AirQuality = pd.read_csv("../DS320A_2023SP/data/AirQualityUCI.csv", sep = ';')
print(df_AirQuality.shape)
```

```python
df_AirQuality.head()
```

```python
#---prepare data to plot line
#---compute the mean of NMHC per day

data = df_AirQuality[~(df_AirQuality['NMHC(GT)'] == -200)]

data = data[['Date','NMHC(GT)']]

data['Date'] = pd.to_datetime(data['Date']) 

data = data.groupby('Date').agg({'NMHC(GT)': 'mean'})
data = data.reset_index()

data = data.sort_values(['Date'], ascending = True )
print(data.shape)
data.head()
```

```python
#--- annotation preparation
#--- we want to annotate 2 days with highest NMHC
annos= data['NMHC(GT)'].nlargest(n = 2)
print(annos)
```

```python
#---ploting just data
freq = 10
x_ticks =  np.arange(start = 1, stop = data.shape[0]+1, step = freq)
x_ticklabels =  data.Date[x_ticks]
x_ticklabels = x_ticklabels.dt.strftime('%Y-%m-%d')

fig = plt.figure(figsize = (8,5))
ax = fig.add_subplot()
ax.plot(range(data.shape[0]), data['NMHC(GT)'])

ax.set_title("Italy, 03/2004 - 02/2005, Non Metanic Hydrocarbons level")
ax.set_xlabel("Day")
ax.set_ylabel("NMHC average ($microg / m^3$)")

ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels, rotation = 45)


```

```python
#---ploting data + annotation

#---ploting the line
fig, ax = plt.subplots(figsize = (9,5)) #plt.figure(figsize = (7,5))
#ax = fig.add_subplot()
ax.plot(range(data.shape[0]), data['NMHC(GT)'])

ax.set_title("Italy, 03/2004 - 02/2005, Non Metanic Hydrocarbons level")
ax.set_xlabel("Day")
ax.set_ylabel("NMHC average ($microg / m^3$ )")

ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels, rotation = 45)
ax.set_ylim([20, 700])

#---ploting annotation
for i in range(len(annos)):
    ax.annotate(text = data['Date'][annos.index[i]].strftime('%Y-%m-%d'), 
                xy=(annos.index[i], data['NMHC(GT)'][annos.index[i]] + 20),
                xytext=(annos.index[i], data['NMHC(GT)'][annos.index[i]] + 70),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left", verticalalignment="top")

#---here: xy = the coordinates of the points you want to annotate
#-------  xytext: the coordinates you want to put text at
```

```python
#7 scatter: color and size
#--- visit this link for the iris dataset description: https://archive.ics.uci.edu/ml/datasets/iris
#--- we will scatter data points in the iris dataset with 3 colors coresponding 3 groups of data points
df_iris = pd.read_csv("../DS320A_2023SP/data/iris.csv")
df_iris
```

```python
#---we add one column of numerical values represent the groups
#---Iris-setosa <=> 1; Iris-versicolor <=> 2; Iris-virginica = 3
def get_groupID_by_name(name):
    if name == 'Iris-setosa':
        return 1
    elif name == 'Iris-versicolor':
        return 2
    else:
        return 3
    
df_iris['group'] = df_iris['Name'].apply(get_groupID_by_name)
df_iris
```

```python
#---let's scatter data points and their centroids

centroids = df_iris.groupby('group').agg({'SepalWidth': 'mean', 'SepalLength': 'mean', 'PetalWidth': 'mean', 'PetalLength': 'mean'})
centroids = centroids.reset_index()

fig, ax = plt.subplots(figsize = (7,4))

ax.scatter(x = df_iris['PetalLength'], y = df_iris['PetalWidth'], s=20, c=df_iris['group'], alpha=0.3, cmap = 'jet')
ax.scatter(x = centroids['PetalLength'], y = centroids['PetalWidth'], s=100, c=centroids['group'], alpha=0.7, cmap = 'jet')
ax.set_title("Iris data visualization")
ax.set_xlabel("PetalLength (cm)")
ax.set_ylabel("PetalWidth (cm)")


```
