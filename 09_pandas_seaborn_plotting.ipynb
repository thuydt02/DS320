Matplotlib provides low-level tools for you to visualize data, for example, you need to assemble lines, bars, scatters,...

Pandas itself provides high-level tools for data visualization. Recall that Pandas has 3 structures to store data, including DataFrame, Series, and Index. Here DataFrame and Series have method for plotting

Seaborn is based on matplotlib. Like Pandas, it wrappes low-level tools in methods to simplify your work of making plots.

# Plotting with pandas

Series and dataframe have the `plot` attribute to make plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#1 line plots with series
#---by default, the plot attribute will plot line(s) from a Series/dataframe
#---steps:
#------[1] create a figure containing plots. 
#------    Recall that in a figure you can add subplots
#------[2] data preparation
#------[3] plotting

fig = plt.figure(figsize = (4,3))
ax = fig.add_subplot()

x = np.arange(start = -1, stop = 1 + 0.1, step = 0.1)
s = pd.Series(data = x ** 2, index = x)

s.plot(kind = 'line', ax = ax, title = 'Seires.plot()', label = '$y = x^2$', 
       alpha = 0.5, color= 'red',
       xlabel = 'x', ylabel = 'y',
       grid = False,
       legend = True)

```

```python
#1 your turn:
#---plot 2 Series s1 = x^3 and s2 = x^4 in two subplots here x in [-2, 2]
fig = plt.figure(figsize = (4,3))
ax = fig.add_subplot()

x = np.arange(start = -2, stop = 2 + 0.1, step = 0.1)
s1 = pd.Series(data = x ** 3, index = x)
s2 = pd.Series(data = x ** 4, index = x)
s1.plot(kind = 'line', ax = ax, title = 'Seires.plot()', label = '$y = x^3$', 
       alpha = 0.5, color= 'blue',
       xlabel = 'x', ylabel = 'y',
       grid = False,
       legend = True)
s2.plot(kind = 'line', ax = ax, title = 'Seires.plot()', label = '$y = x^4$', 
       alpha = 0.5, color= 'red',
       xlabel = 'x', ylabel = 'y',
       grid = False,
       legend = True)
```

```python
#1 line plots with series
#--- use the ax param to add objects into a plot

fig = plt.figure(figsize = (4,3))
ax = fig.add_subplot()

x = np.arange(start = -1, stop = 1 + 0.1, step = 0.1)
s = pd.Series(data = x ** 2, index = x)

s.plot(kind = 'line', ax = ax, title = 'Seires.plot()', label = '$y = x^2$', 
       alpha = 0.5, color= 'red',
       xlabel = 'x', ylabel = 'y',
       grid = False,
       legend = True,
       ylim = [0, 1.5])

annos = [0.5, 1]
for i in range(len(annos)):
    ax.annotate(text = annos[i] ** 2, 
                xy=(annos[i], annos[i]**2 + 0.05),
                xytext=(annos[i], annos[i]**2 + 0.3),
                arrowprops=dict(facecolor="black",alpha = 0.5,
                                headwidth=4, width=2, headlength=4),
                horizontalalignment="left", verticalalignment="top")
```

```python
#1 line plots with dataframe
#--- we plot one line for one column

fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot()

x = np.arange(start = 0, stop = math.pi * 3, step = 0.1)

s1 = pd.Series(data = np.sin(x), index = x)
s2 = pd.Series(data = np.cos(x), index = x)

df = pd.DataFrame({"sin": s1, "cosine": s2})

df.plot(kind = 'line', ax = ax, title = 'DataFrame.plot()', 
       color = ['red', 'orange'],  alpha = 0.7,
       xlabel = 'x', ylabel = 'y',
       grid = False,
       legend = True)
```

```python
#1 line plots with dataframe
#--- your turn: plot 2 lines of the following df: x^3 and tanh
#-------        use column x for xtick

fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot()

x = np.arange(start = -1, stop = 1 + 0.1, step = 0.1)

sigmoid = pd.Series(data = 1 / (1 + np.exp(-x)), index = x)

softplus = pd.Series(data = np.log(np.exp(x) + 1), index = x)

tanh = pd.Series(data = np.tanh(x), index = x)

x3 = pd.Series(data = x**3, index = x)

df = pd.DataFrame({"x": x, "sigmoid": sigmoid, "softplus": softplus, 
                   "tanh": tanh, "x3": x3})


df[['x3','tanh']].plot(kind = 'line', ax = ax, title = 'DataFrame.plot()', 
       color = ['red', 'orange'],  alpha = 0.7,
       xlabel = 'x', ylabel = 'y',
       xticks = df['x'], grid = False,
       legend = True)

```

```python
#2 bar plots with series

df = pd.read_csv('./data/continents.csv')
df = df.sort_values('Population', ascending = False)

fig, ax = plt.subplots(2, 1, figsize = (6,10))

df['Population'].plot(kind = 'bar', ax = ax[0], color="teal", alpha = 0.8,
                     title = 'Vertical Bar Plot');

ax[0].set_xticklabels(df['Continent'], rotation = 45)

df['Population'].plot(kind = 'barh', ax = ax[1], color="teal", alpha = 0.8,
                     title = 'Horizontal Bar Plot');

ax[1].set_yticklabels(df['Continent'], rotation = 0)
```

```python
#2 bar plots: stacked

df_tips = pd.read_csv("../DS320A_2023SP/data/tips.csv")
df_tips
```

```python
party_counts = pd.crosstab(df_tips["day"], df_tips["size"])
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"])
party_counts
```

```python
#---remove column 1 and 6 as few parties
#---normalize data such that each row sums up to 1 

party_counts = party_counts.loc[:, 2:5]
party_percents = party_counts.div(party_counts.sum(axis="columns"), axis="index")
party_percents
print(party_percents.columns)
```

```python
#---bar plot
fig = plt.figure(figsize = (5,3))
ax = fig.add_subplot()

party_percents.plot(kind = "bar", stacked = True, ax = ax, 
                    color = {2: "red", 3: "orange", 
                            4: "yellow", 5: "pink"},
                    alpha = 0.5, width = .5,
                    title = 'DataFrame: Stacked Bar Plot');
```

```python
#2: bar plots:
#--- your turn
#---create bar plot to compare total tips of days.
#---x-axis = {Thur, Fri, Sat, Sun}
#---y-axis = {sum of tips for Thurs, ..., sum of tips for Suns}


def day2number(day):
    d = day.lower()
    if d == "mon":
        return 1
    elif d == 'tue':
        return 2
    elif d == 'wed':
        return 3
    elif d == 'thur':
        return 4
    elif d == 'fri':
        return 5
    elif d == 'sat':
        return 6
    else:
        return 7

    
data = df_tips.groupby('day').agg({'tip':'sum'})
data = data.reset_index()
data['day_order'] = data.day.apply(day2number)
data = data.sort_values('day_order')


fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot()

data['tip'].plot(kind = "bar", ax = ax, color = 'red',
            alpha = 0.5, width = .5,
            title = 'Tips comparision',
            );
ax.set_xticklabels(data['day'])

#df_tips
```

```python
#3 histogram and density
df_tips["tip_pct"] = df_tips["tip"] / (df_tips["total_bill"] - df_tips["tip"])
df_tips.head()

fig = plt.figure(figsize = (5,3))
ax = fig.add_subplot()

df_tips["tip_pct"].plot(kind = 'hist', ax = ax, bins=50, color = 'red', alpha = 0.5)
```

```python
#4 scatter
fig = plt.figure(figsize = (5,3))
ax = fig.add_subplot()
df_tips.plot(kind = 'scatter', ax = ax, x="tip", y="total_bill",
           color = 'red', alpha = 0.5, s = 20)
```

```python
#1 lines

import seaborn

fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot()

x = np.arange(start = 0, stop = math.pi * 3, step = 0.1)

s1 = pd.Series(data = np.sin(x), index = x)
s2 = pd.Series(data = np.cos(x), index = x)

df = pd.DataFrame({"x": x, "sin": s1, "cosine": s2})

seaborn.lineplot(data = df[["x", "sin"]], ax = ax, x = "x", y = "sin", 
       color = 'red', alpha = 0.7, legend = 'auto', label = 'sin')

seaborn.lineplot(data = df[["x", "cosine"]], ax = ax, x = "x", y = "cosine", 
       color = 'green', alpha = 0.7, legend = 'auto', label = 'cosine')

ax.set_ylabel("y")
```
