```python
#tuple

a, b = (1, 2,3, 'aa', 'bb', 'cc'), ('aa', 'bb', 'cc', 'dd')

#1 accessing: indexing and slicing

print(a[1])
print(a[4:6])
print(a[:5]) 
print(a[5:])

#2 concatenate
print(a + b)

#3 repetition
print(a * 2)

#4 membership
for e in a:
    print(e)

#5 conversion: tuple -> list | set | string | dict
print(list(a))
print(set(a + b))
c = ((1, 'aaa'), (2, 'bbb'), (3, 'ccc'))
print(dict(c))

#0 immutable
#a[0] = 345 # does not work
```

```python
#list

a, b = [1, 2,3, 'aa', 'bb', 'cc'], ['aa', 'bb', 'cc', 'dd']

#1 accessing: indexing and slicing

print(a[1])
print(a[4:6])
print(a[:5]) 
print(a[5:])

#2 concatenate
print(a + b)

#3 repetition
print(a * 2)

#4 membership
for e in a:
    print(e)

#5 conversion: list -> tuple | set | string | dict
print(tuple(a))
print(set(a + b))
c = [[1, 'aaa'], [2, 'bbb'], [3, 'ccc']]
print(dict(c))

#0 immutable
#a[0] = 345 # does not work
```

```python
#set

a, b = {1, 2,3, 'aa', 'bb', 'cc'}, {'aa', 'bb', 'cc', 'dd'}

#1 accessing: indexing and slicing, not applicable as sets are unordered

#2 concatenate
print(a.union(b))

#3 repetition, not applicable

#4 membership
for e in a:
    print(e)

#5 conversion: set -> tuple | list | string | dict
print(tuple(a))
print(list(a))
c = {(1, 'aaa'), (2, 'bbb'), (3, 'ccc')}
print(dict(c))

#0 immutable
#a[0] = 345 # does not work
```

```python
#string

a, b = "123abc", 'abcd'

#1 accessing: indexing and slicing

print(a[1])
print(a[4:6])
print(a[:5]) 
print(a[5:])

#2 concatenate
print(a + b)

#3 repetition
print(a * 2)

#4 membership
for e in a:
    print(e)

#5 conversion: string -> tuple | set | list | dict
print(tuple(a))
print(set(a))
print(list(a))

#0 immutable
#a[0] = 345 # does not work

# more on strings: methods
#--- get words from a string
s = "abc,def,123"
words = s.split(',')
print(words)

#--- join elements in a list => a string
s1 = ",".join(words)
print(s1)

#--- review find(), replace(), ... at home
```

```python
#dict

a, b = {1: 'aa', 2: 'bb', 3:'cc', 4:'dd'}, {1: 'aa', 2: 'bb', 5: 'ee'}
a[6] = 'st'
#1 accessing: indexing

print(a[1])

#2 concatenate
a.update(b)
print(a)  #a = {}

#3 repetition
print(a * 2)

#4 membership
for e in a:
    print(e)

#5 conversion: list -> tuple | set | string | dict
print(tuple(a))
print(set(a + b))
c = [[1, 'aaa'], [2, 'bbb'], [3, 'ccc']]
print(dict(c))

#0 immutable
#a[0] = 345 # does not work
```
